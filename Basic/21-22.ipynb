Tính tổng và tích

```python
import numpy as np

arr = np.array([[1, 2, 3],[4, 5, 6]])
tong = 0
tich = 1

for i in arr:
    for j in i:
        tong = tong+j
        tich = tich*j

print(tong)
print(tich)
```

Cho một dãy số tự nhiên 20 phần tử, hãy thay thế tất cả những phần tử lẻ bằng số -1

```python
# Bước 1: Tạo dãy số tự nhiên gồm 20 phần tử
sequence = [i for i in range(1, 21)]

# Bước 2 và 3: Duyệt qua từng phần tử trong dãy số và kiểm tra số lẻ
for i in range(len(sequence)):
    if sequence[i] % 2 != 0:  # Kiểm tra số lẻ
        sequence[i] = -1  # Thay thế bằng -1

# Bước 4: In ra dãy số đã được thay thế
print(sequence)
```

Tạo một ma trận 3x5 gồm các số ngẫu nhiên từ 0 đến nhỏ hơn 10, tính và in ra số lớn
nhất trên mỗi dòng + mỗi cột của ma trận

```python
import numpy as np

# Bước 1: Tạo ma trận 3x5 gồm các số ngẫu nhiên từ 0 đến nhỏ hơn 10
matrix = np.random.randint(0, 10, (3, 5))

# Bước 2: Tính số lớn nhất trên mỗi cột của ma trận
max_in_cols = np.max(matrix, axis=0)
max_in_rows = np.max(matrix, axis=1)

# Bước 3: In ma trận và số lớn nhất trên mỗi cột
print("Ma trận 3x5:")
print(matrix)

print("\nSố lớn nhất trên mỗi cột:")
print(max_in_cols)

print("\nSố lớn nhất trên mỗi dòngt:")
print(max_in_rows)
```

Vẽ biểu đồ y = x^2

```python
import matplotlib.pyplot as plt
import numpy as np

# Tạo một dãy giá trị x từ -10 đến 10
x = np.linspace(-20, 20, 400)

# Tính giá trị y tương ứng với mỗi giá trị x
y = x**2

# Vẽ đồ thị
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = x^2")

# Đặt tiêu đề và nhãn cho các trục
plt.title("Đồ thị của hàm số y = x^2")
plt.xlabel("x")
plt.ylabel("y")

# Hiển thị lưới và chú thích
plt.grid(True)
plt.legend()

# Hiển thị đồ thị
plt.show()
```
